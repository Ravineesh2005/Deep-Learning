```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('practice..csv')
```

```python
plt.scatter(x=df['input_1'],y=df['input_2'],c=df['output'])
```

```python

X = df.iloc[:100,0:2].values
y = df.iloc[:100,-1].values
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Input
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
model_1 = Sequential([
    Input(shape=(2,)),
    Dense(5,activation='sigmoid'),
    Dense(5,activation='sigmoid'),
    Dense(1,activation='sigmoid')
])
# using sigmoid function in all hidden layers and arctucture is [2'5'5'1]
```

```python
initial_weights = model_1.get_weights()
```

```python
initial_weights[0] = np.ones(model_1.get_weights()[0].shape)
initial_weights[1] = np.ones(model_1.get_weights()[1].shape)
initial_weights[2] = np.ones(model_1.get_weights()[2].shape)
initial_weights[3] = np.ones(model_1.get_weights()[3].shape)
initial_weights[4] = np.ones(model_1.get_weights()[4].shape)
initial_weights[5] = np.ones(model_1.get_weights()[5].shape)
```

```python
initial_weights
```

```python
model_1.set_weights(initial_weights)
```

```python
model_1.get_weights()
```

```python
model_1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

history = model_1.fit(X,y,epochs=10,validation_split=0.2)

plot_decision_regions(X,y.astype('int'), clf=model_1, legend=2)
```

```python
model_1.get_weights()

#the arcutracure [2,5,5,1] behaves like [2,1,1,1]
```
